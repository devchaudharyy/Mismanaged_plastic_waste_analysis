# Mismanaged Plastic Waste Analysis ( WORLD-WIDE)

Mismanaged plastic waste is plastic that is either littered or inadequately disposed. A country's total does not
include waste that is exported overseas, where it may be mismanaged.

In this notebook, we are going to analyse a dataset conataing data of mismanaged plastic waste produced by **countries all over world** with an additional data of mismanaged plastic waste produced **per capita ( country-wise )** of the year 2019 and 2010.

We will compare the 2019 and 2010 data on different aspects and will learn about the trends of each year. We will use different visualisation tools, like **bar chart, pie chart**, to interpret the data more easily and get deep knowledge.

### Different libraries worked on :-
* Pandas
* Numpy
* Matplotlib.pyplot

```python
# importing different libraries in Python in order to use all the built-in functios.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
# reading csv file into a dataframe using pandas

df = pd.read_csv('mismanaged_plasticwaste.csv')
```

```python
# displaying top 5 rows of the dataframe.
 
df.head(5)
```

```python
# renaming column names for easy and quick access

df = df.rename( columns = {'Total_MismanagedPlasticWaste_2010 (millionT)' : 'Total_mT_2010',
                           'Total_MismanagedPlasticWaste_2019 (millionT)' : 'Total_mT_2019',
                          'Mismanaged_PlasticWaste_PerCapita_2010 (kg per year) ':'PerHead_kg_2010',
                      'Mismanaged_PlasticWaste_PerCapita_2019 (kg per year) ':'PerHead_kg_2019'})
```

```python
df.head(5)
```

```python
# learning more about the dataframe
df.shape
```

```python
#getting statistical data 
df.describe()
```

```python
df.info()
```

```python
# displaying maximum amount of mismanaged plastic waste produced by a country over all and per Capita in 2010

print("Maximum mismanaged waste produced in 2010 by a country :", df['Total_mT_2010'].max(), "millionT")
print("Maximum mismanaged waste produced in 2010 per Capita : ",df['PerHead_kg_2010'].max(), "Kg")
```

```python
# displaying maximum amount of mismanaged plastic waste produced by a country over all and per Capita in 2019

print("Maximum mismanaged waste produced in 2019 by a country : ",df['Total_mT_2019'].max(), "millionT")
print("Maximum mismanaged waste produced in 2019 per Capita : ",df['PerHead_kg_2019'].max(), 'Kg')
```

```python
# displaying the rise in mismanaged plastic waste production from year 2010 to year 2019

print("Increase in mismanaged plastic waste from 2010 to 2019 (country-wise) : ",
       df['Total_mT_2019'].sum() - df['Total_mT_2010'].sum(), "millionT" )
```

```python
# displaying the rise in mismanaged plastic waste production per Capita from 2010 to 2019

print("Increase in mismanaged plastic waste from 2010 to 2019 (per-capita) : ", 
       df['PerHead_kg_2019'].sum() - df['PerHead_kg_2010'].sum(), "kg" )
```

```python
# displaying average waste produced by countries in 2019 and 2010

print("Average waste produced by countries in 2019 : ", 
       df['Total_mT_2019'].mean() , "millionT" )

print("Average waste produced by countries in 2010 : ", 
       df['Total_mT_2010'].mean() , "millionT" )
```

```python
# displaying average waste produced bt countries per capita in 2019 and 2010

print("Average waste produced by countries per capita in 2019 : ", 
       df['PerHead_kg_2019'].mean() , "KG per year" )

print("Average waste produced by countries per capita in 2010 : ", 
       df['PerHead_kg_2010'].mean() , "KG per year" )
```

**We've Analysed the basic trends from the dataframe in 2010 and 2019, now we will visualise the outcomes and other trends with the help of visualisation tools for better understanding.**

# VISUALISING DATA

```python
x = ['2019','2010']
y = [df['Total_mT_2019'].sum(),df['Total_mT_2010'].sum()]
plt.stem(x,y, linefmt = "-")

font1 = {'family':'serif','color':'darkred','size':15}
font2 = {'family':'serif','color':'black','size':10}

plt.ylabel("WASTE IN millionT", fontdict = font2, weight = "bold")
plt.xlabel("YEAR" ,fontdict = font2, weight = "bold")

plt.title("Total mismanaged plastic waste in 2019 and 2010", fontdict= font1, pad = 20, weight = "bold")

plt.show()
```

A huge rise in mismanaged plastic waste production can be observed in the 2019 as compared to 2010

```python
x = (2019,2010)
y = (df['PerHead_kg_2019'].sum(),df['PerHead_kg_2010'].sum())
plt.stem(x,y, linefmt = "-")

font1 = {'family':'serif','color':'darkred','size':15}
font2 = {'family':'serif','color':'black','size':10}

plt.ylabel("WASTE IN KG PER YEAR", fontdict = font2, weight = "bold")
plt.xlabel("YEAR" ,fontdict = font2, weight = "bold")

plt.title("Total mismanaged plastic waste per capita in 2010 and 2019", fontdict= font1, pad = 20, weight = "bold")


plt.show() 
```

**A massive escalation has been observed in both total country waste and per capita waste in 2019 which is a concerning topic and need to looked after immediately.The main hotspots of mismanaged plastic waste generation would gradually move from *Asia to Africa***

There are various reasons responsible for this escalation and they are -
* Uncontrolled landfills
* Littering
* Open burning
* Ocean dumping

# How can we manage our plastic waste?

* Recycle when possible (do it properly)
* Avoid single-use plastics
* Use alternative packaging 
* Do a trash audit
* Find reusable options
* Choose clothes made from sustainable and recycled materials

**If we really want to make headway on this issue, we need a Global Treaty on plastic pollution as a fundamental building block for standardized measurement to inform policymakers, corporate decision makers, and the public.**
